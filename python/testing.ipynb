```python
%matplotlib notebook
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from thomson import *
from lsst.sims.utils import _healbin
import healpy as hp
from lsst.sims.featureScheduler import read_fields
```

```python
npts = 500
theta, phi = fib_sphere_grid(npts)
# Note, declination = phi - pi/2, RA = theta
x, y, z = thetaphi2xyz(theta, phi)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z)
```

```python
nside = 16
density = _healbin(theta, phi-np.pi/2, theta*0+1, nside=nside, reduceFunc=np.sum)
hp.mollview(density, unit='Number of points', title='Fibonacci Spiral')
```

```python
indx=200
xn,yn,zn = minimize_single(x,y,z, indx)
```
